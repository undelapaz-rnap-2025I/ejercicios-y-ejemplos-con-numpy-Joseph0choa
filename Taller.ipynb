```python
x = np.array([1,2,3,4,5]).reshape(5,1)
y = np.array([3,6,8,11,13]).reshape(5,1)


def cost_function(theta, x, y):
    m = len(y)
    h = X @ theta
    E = y - h
    E_square = E.T @ E
    J = 1/(2*m)*E_square
    return J
```

Gradient Descent Optimization

```python
iterations = 1000
m = len(y)
X = np.hstack([np.ones((m, 1)), x])
theta = np.zeros((2, 1))
alpha = 0.01

theta_progress = []
error_progress = []

def gradientDescent(X, y, theta, alpha, iterations):
  m = len(y) 
  J_history = np.zeros(iterations)

  for iter in range(iterations):

      constant = (alpha/m)
      h = X @ theta
      E = h - y
      derivate_J = X.T @ E
      theta = theta - constant * derivate_J
        
      theta_progress.append(theta.ravel().tolist())
      error_progress.append(np.sum(E**2) / (2 * m))
      

      J_history[iter] = cost_function(theta,X, y)
      print(f"Iteration {iter + 1}\tCost: {J_history[iter]:.6f}")
  return theta

theta = gradientDescent(X, y, theta, alpha, iterations);
print("Theta found by gradient descent: ", theta.ravel())

```

 Explain in your own words the purpose of the learning rate in gradient
 descent optimization.

The purpose of the learning rate in gradient descent optimization is to define a measure that determines how quickly or slowly the system can reach an optimal point—that is, to minimize the cost function values. Properly adjusting this value is crucial, as it directly impacts the computational cost and the stability of the system, potentially causing the model to diverge if it is not chosen appropriately.
